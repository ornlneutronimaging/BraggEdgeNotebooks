```python
import time
import datetime
def dateparse (date_time):
    return time.mktime(datetime.datetime.strptime(date_time, '%m/%d/%Y %I:%M:%S %p').timetuple())
```

```python
#df = pd.read_csv(output_file, header=None, names=['time', 'temperature (C)'])
df = pd.read_csv(output_file, parse_dates=True, date_parser=dateparse, header=None)
```
